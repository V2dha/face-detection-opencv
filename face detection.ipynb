```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img= plt.imread("group_picture.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
detector
```

```python
faces=detector.detectMultiScale(img)
```

```python
faces
```

```python
for face in faces:
    x,y,w,h=face
    
    modified_img=cv2.rectangle(img, (x,y), (x+w,y+h), (100,230,255),2)
```
